```python
def n_puntos_imagen(n,image_to_test):
    data = np.asarray(image_to_test)
    data_custom = data.copy()
    data_custom.flags.writeable = True
    contador = 0
    for i in range(n):
        x = random.randint(0,len(data)-1)
        y = random.randint(0,len(data[x])-1)
        if (data[x][y].tolist() == [255, 0, 255]):
            contador = contador + 1
        data_custom[x][y] = [0,0,0]
    image = Image.fromarray(data_custom)
    display(image)
    return contador/n
```

```python
n_puntos_imagen(1000000,image)
```
